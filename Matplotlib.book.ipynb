## Location of .legeng()
***

```python
# .loc - location of a legend
locs = ['best', 'upper right', 'upper left', 'lower left', 'lower right', 'right',\
       'center left', 'center right', 'lower center', 'upper center', 'center']

plt.figure(figsize=(12,12))
for i in range(3):
    for j in range(4):
        if i*4+j < 11:
            # Here we state a location of our diagram
            # 3 - quantity of rows
            # 4 - quantity of columns
            # i*4+j - a place where we want to locate our diagram
            plt.subplot(3, 4, i*4+j+1)
            plt.title(locs[i*4+j]) # a title of a diagram from the list
            plt.plot(x, y1, 'o-r', label='line 1') 
            plt.plot(x, y2, 'o-.g', label='line 2')
            plt.legend(loc=locs[i*4+j])
        else:
            break
```

```python
# For more flexiable managing of a location of our legend, we can use an optional parametr - 
# bbox_to_anchor of .legend()
# For this parametr we can state a tuple for 2 or 4 elements
# bbox_to_anchor = (x, y, width, heght)
# bbox_to_anchor = (x, y)

plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.g', label='line 2')
plt.legend(bbox_to_anchor=(1, .6))
```

```python
# Optional parametrs for .legend()
# fontsize - int or float or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
# frameon - True or False - a frame of a legend 
# framealpha - None or float - a transparency of a legend
# facecolor - None or str - painting of a legend
# edgecolor - None or str - painting of edges
# title
# title_fontsize

plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.m', label='line 2')
plt.legend(fontsize=14, shadow=True, framealpha=1, facecolor='y', edgecolor='g', title='Legend')
```

### Style of an arrow in annotation
***
arrowprops takes dict with values from class FancyArrowPacth

- **Curve**:\
    name: - \
    params: None
- **CurveB**:\
    name: ->\
    params: head_length=.4, head_width=.2
- **BracketB**:\
    name: -\[\
    params: widthB:1.0, lengthB=.2, angleB=None
- **CurveFilledB**:\
    name: -|>\
    params: head_length=.4, head_width=.2
- **CurveA**:\
    name: <-\
    params: head_length=.4, head_width=.2
- **CurveAB**:\
    name: <->\
    params: head_length=.4, head_width=.2
- **CurveFilledA**:\
    name: <|-\
    params: head_length=.4, head_width=.2
- **CurveFilledAB**:\
    name: <|-|>\
    params: head_length=.4, head_width=.2
- **BracketA**:\
    name: ]-\
    params: widthA:1.0, lengthA=.2, angleA=None
- **BracketAB**:\
    name: ]-[\
    params: widthA:1.0, lengthA=.2, angleA=None, widthB:1.0, lengthB=.2, angleB=None
- **Fancy**:
    name: fancy\
    params: head_length=.4, head_width=.2, tail_width=.4
- **Simple**:\
    name: simple\
    params: head_length=.4, head_width=.2, tail_width=.2
- **Wedge**:\
    name: wedge\
    params: tail_width=.3, shrink_factor=.5

```python
plt.figure(figsize=(7, 5))
arrows = ['-', '->', '-[', '|-|', '-|>', '<-', '<->', '<|-', '<|-|>', 'fancy', 'simple', 'wedge']

bbox_properties = dict(boxstyle='round, pad=.2', ec='k', fc='w', ls='-', lw=1)

ofs_x = 0
ofs_y = 0

for i, ar in enumerate(arrows):
    if i == 6:
        ofs_x = .5
        
    plt.annotate(ar, xy=(.4+ofs_x, .92-ofs_y), xycoords='data', xytext=(.05+ofs_x, .9-ofs_y),\
                    textcoords='data', fontsize=17, bbox=bbox_properties, arrowprops=dict(arrowstyle=ar))
    if ofs_y == .75:
        ofs_y = 0
    else:
        ofs_y += .15
```

### Style of a connection of a line
***
connectionstyle

We can state a style of line, which connects dots (xy, xycoords). 
As a value a parametr connectionstyle takes an object of a class ConnectionStyle or string with parametrs, which listed and separated by comma.

- **Angle**:\
    name: angle\

## General setting up
***

One way is to create an Axes element and modify a part of params.
It will be more convinience to use **inset_axes()** from **mpl_tollkits.axes_grid1.inset_locator**.
***

**Parametrs of a function inset_axes()**

- **parent_axes**:\
    type: axes\
    description: an object of master Axes
- **width**:\
    type: float or str\
    description: a width of an object, states in % from a master Axes or as a number
- **height**:\
    type: float or str\
    description: a width of an object, states in % from a master Axes or as a number
- **loc**:\
    type: int or string, optional (a default value is 1)
    description: a location of an object. Takes:\
                - 'upper right': 1
                - 'upper left': 2
                - 'lower left': 3
                - 'lower right': 4
                - 'right': 5
                - 'center left': 6
                - 'center right': 7
                - 'lower center': 8
                - 'upper center': 9
                - 'center': 10
- **bbox_to_anchor**:\
    type: tuple or matplotlib.transforms.BboxBase or optional\
    description: position and correlation of sides of an object. Format: left angle, bottom angle, width, length.
- **bbox_transform**:\
    type: matplotlib.transforms.Transform or optional\
    description: transformation of an object
- **borderpad**:\
    type: float or optional\
    description: backlash between bbox_to_anchor and an object

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
np.random.seed(123)
vals = np.random.randint(11, size=(7,7))
fig, ax = plt.subplots()
gr = ax.pcolor(vals)
axins = inset_axes(ax, width='7%', height='50%', loc='lower left',\
                  bbox_to_anchor=(1.05, .0, 1, 1), bbox_transform=ax.transAxes, borderpad=0)
plt.colorbar(gr, cax=axins)

# If you need, it's possible to modify a scale of a color line by class Tick.
```

For state your own scale you need to give a list to colorbar() function by a param `tick`.

### Step-diagram

<pre>
step(x, y, [fmt], *, data=None, where='pre', **kwargs)
</pre>

 **x**: array\
 **y**: array\
 **fmt**: str, optional:
  - a format of a line
 
 **data**: index object, optional:
  - tags
 
 **where**: {'pre', 'post', 'mid'}, optional:
 - defines a place where will be stand a step
  - 'pre' - a value of y[i] defines for (x[i-1], x[i])
  - 'post' - a value of y[i] defines for (x[i], x[i+1])
  - 'mid' - a value of y stands in the middle of an interval

```python
x = np.arange(0, 7)
y = x

where_set = ['pre', 'post', 'mid']
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

for i, ax in enumerate(axs):
    ax.step(x, y, 'g-o', where=where_set[i])
    ax.grid()
    ax.set_title(where_set[i])
```

### Stack-diagram

For stack diagram use **stackplot()**. Diagrams drawings like one after another and each following diagram is a sum of previous one and targeted.

```python
x = np.arange(0, 11, 1)
y1 = np.array([(-.2)*i**2+2*i for i in x])
y2 = np.array([(-.4)*i**2+4*i for i in x])
y3 = np.array([2*i for i in x])

labels = ['y1', 'y2', 'y3']
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels = labels)
ax.legend(loc='upper left')
```

Top edge for y2 equals as sum of values from y1 and y2, for y3 - sum of values from y1, y2 and y3.

#### Additional params for stem()

 **linefmt: str, optional**:

 - '-' - solid line style
 - '--' - dashed line style
 - '-.' - dash-dot line style
 - ':' - dotted line style

 **markerfmt: str, optional**:

 - 'o' - circle

```python
x = np.arange(0, 10.5, .5)
y = np.array([(-.2)*i**2+2*i for i in x])

marker_set = ['o', '+', '*', '.', 'x', 's', 'd', '^', 'v', '<', '>', 'p', 'h', 'none']
fig, axs = plt.subplots(14, 1, figsize=(10, 40))
axs = axs.flatten() # use it because of https://issue.life/questions/45769005

for i, axis in enumerate(axs):
    axis.stem(x, y, linefmt='r--', markerfmt=marker_set[i], bottom=.3, use_line_collection=True)
    axis.grid()
    axis.set_title(marker_set[i])

plt.tight_layout()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)

# constrained_layput - https://matplotlib.org/3.1.0/tutorials/intermediate/constrainedlayout_guide.html

p1 = axs[0].imshow(data, cmap='winter', aspect='equal', vmin=-1, vmax=1, origin='lower')
fig.colorbar(p1, ax=axs[0])

p2 = axs[1].imshow(data, cmap='plasma', aspect='equal', interpolation='gaussian', origin='lower', extent=(0, 30, 0, 30))
fig.colorbar(p2, ax=axs[1])
```
